# Question 1

```python
import numpy as np # importing numpy library
import pandas as pd # importing pandas library
import warnings # importing warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv('train.csv') # reading train dataset and storing it as train
test = pd.read_csv('test.csv') # reading test dataset and storing it as test
```

```python
train['Sex'] = train['Sex'].replace(["female", "male"], [1, 0]) # replacing female and male as 1 and 0 respectively
train['Embarked'] = train['Embarked'].replace(['S','C','Q'],[0,1,2]) # replacing S,C,Q with 0,1,2 respectively
```

```python
train.fillna(train.mean(),axis=0,inplace=True) # filling all the NaN values with mean of respective columns
```

### 3. Implement Naïve Bayes method using scikit-learn library and report the accuracy.

```python
X = train.drop(columns = ['Name','Survived','Ticket','Cabin']) # considering only the columns which are of numeric type
Y = train[['Survived']] # considering the target column
```

```python
from sklearn.metrics import confusion_matrix # importing confusion_matrix function
from sklearn.metrics import classification_report # importing classification_report function
from sklearn.model_selection import train_test_split # importing train_test_split function
```

```python
# spliting the given dataset into train and test datasets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB # importing GaussianNB 
classifier = GaussianNB() # creating the model
classifier.fit(X_train, Y_train) # feeding model with training dataset
Y_pred = classifier.predict(X_test) # predicting the dependent variable in the test dataset
# Summary of the predictions made by the classifier
print(classification_report(Y_test, Y_pred)) # printing the classification report
print(confusion_matrix(Y_test, Y_pred)) # printing the confusion matrix
# Accuracy score
from sklearn.metrics import accuracy_score # importing the accuracy_score
print('accuracy is',accuracy_score(Y_pred,Y_test)) # printing the accuracy of predicted values with true values
```
